```python
import numpy as np
import matplotlib.pyplot as plt

# sizes
N_inp, N_exc, N_inh = 200, 432, 108
N = N_inp + N_exc + N_inh

# RNG (one place to change the seed)
rng = np.random.default_rng(42)
```

```python
# 0=inp, 1=exc, 2=inh
types = np.r_[np.zeros(N_inp, int), np.ones(N_exc, int), np.full(N_inh, 2, int)]
is_inp = (types == 0)
is_exc = (types == 1)
is_inh = (types == 2)
idx_net = np.where(~is_inp)[0]  # indices of E+I neurons

# masks (pre on rows, post on cols)
masks = {
    "E->E":   is_exc[:, None] & is_exc[None, :],
    "E->I":   is_exc[:, None] & is_inh[None, :],
    "I->E":   is_inh[:, None] & is_exc[None, :],
    "I->I":   is_inh[:, None] & is_inh[None, :],
    "inp->E": is_inp[:, None] & is_exc[None, :],
}
# remove autapses in same-pop projections
for k in ("E->E", "I->I"):
    np.fill_diagonal(masks[k], False)
```

```python
# 6 x 6 x 15 grid (540 sites) for E+I only
gx, gy, gz = 6, 6, 15
assert gx * gy * gz == (N_exc + N_inh)

grid = np.stack(
    np.meshgrid(np.arange(gx), np.arange(gy), np.arange(gz), indexing='ij'),
    axis=-1
).reshape(-1, 3).astype(float)
grid -= grid.mean(axis=0)     # center
spacing = np.array([1.0, 1.0, 1.0])  # grid units
grid *= spacing

# assign sites: randomized E/I placement (comment two lines below to use “first E then I”)
perm = rng.permutation(grid.shape[0])
grid = grid[perm]

pos = np.zeros((N, 3), float)
pos[is_exc] = grid[:N_exc]
pos[is_inh] = grid[N_exc:]

# pairwise Euclidean distances for network cells
P = pos[idx_net]                        # (540, 3)
diff = P[:, None, :] - P[None, :, :]
D_net = np.sqrt(np.sum(diff * diff, axis=2))  # (540, 540)

D = np.zeros((N, N), float)
D[np.ix_(idx_net, idx_net)] = D_net     # rows/cols for inputs remain 0
```

```python
def gamma_from_mean_cv(rng, mean, cv, size):
    # mean μ, coefficient of variation c=σ/μ  -> shape k=1/c^2, scale θ=μ*c^2
    k = 1.0 / (cv ** 2)
    theta = mean * (cv ** 2)
    return rng.gamma(k, theta, size=size)

def normal_from_mean_cv_pos(rng, mean, cv, size):
    sigma = mean * cv
    x = rng.normal(loc=mean, scale=sigma, size=size)
    bad = x <= 0
    while np.any(bad):
        x[bad] = rng.normal(loc=mean, scale=sigma, size=bad.sum())
        bad = x <= 0
    return x

# connection rules (probabilities); distance-dependent use c and lambda
conn_specs = {
    "inp->E": {"p": 0.50, "mode": "uniform"},
    "E->E":   {"p": 0.50, "mode": "uniform"},
    "E->I":   {"p": 0.04, "mode": "distance", "c": 2e5, "lam": 0.25},
    "I->E":   {"p": 0.05, "mode": "distance", "c": 4e5, "lam": 0.25},
    "I->I":   {"p": 0.04, "mode": "distance", "c": 1e5, "lam": 0.25},
}

# weight and delay specs
weight_specs = {
    "E->E":   (2.5,   0.7,  +1),
    "E->I":   (1000., 0.7,  +1),
    "I->E":   (1375., 0.7,  -1),
    "I->I":   (6000., 0.7,  -1),
    "inp->E": (15.,   0.07, +1),
}
delay_specs = {
    "E->E":   (5.0, 0.5),
    "E->I":   (5.0, 0.5),
    "I->E":   (2.0, 0.5),
    "I->I":   (2.0, 0.5),
    "inp->E": (2.0, 0.5),
}
```

```python
def sample_connections(rng, masks, D, specs, no_autapse=True):
    N = next(iter(masks.values())).shape[0]
    C = np.zeros((N, N), dtype=bool)

    for key, rule in specs.items():
        m = masks[key]
        n = m.sum()
        if n == 0:
            continue
        p_target = rule["p"]

        if rule["mode"] == "uniform":
            C[m] = rng.random(n) < p_target
        else:
            # kernel k = c * exp(-lam * d)
            dsub = D[m]
            k = rule["c"] * np.exp(-dsub/rule["lam"])

            # bounded mapping p = 1 - exp(-alpha * k), choose alpha so mean(p)=p_target
            lo, hi = 0.0, 1e12
            for _ in range(40):
                mid = 0.5 * (lo + hi)
                p_mid = 1.0 - np.exp(-mid * k)
                if p_mid.mean() < p_target:
                    lo = mid
                else:
                    hi = mid
            p_ij = 1.0 - np.exp(-hi * k)
            C[m] = rng.random(n) < p_ij

    if no_autapse:
        np.fill_diagonal(C, False)
    return C

C = sample_connections(rng, masks, D, conn_specs, no_autapse=True)

# quick sanity check on realized connection fractions
for k, m in masks.items():
    realized = C[m].mean() if m.any() else np.nan
    print(f"{k:6s}: realized {realized:.3f}, target {conn_specs[k]['p']:.3f}")
```

```python
W = np.zeros((N, N), float)
delays = np.zeros((N, N), float)

for key, (mean, cv, sign) in weight_specs.items():
    m = masks[key] & C
    n = m.sum()
    if n:
        W[m] = sign * gamma_from_mean_cv(rng, mean, cv, size=n)

for key, (mean, cv) in delay_specs.items():
    m = masks[key] & C
    n = m.sum()
    if n:
        delays[m] = normal_from_mean_cv_pos(rng, mean, cv, size=n)
```

```python
import matplotlib.colors as mcolors

# symmetric log norm for signed weights
nz = np.abs(W[W != 0])
vmax = nz.max() if nz.size else 1.0
linthresh = np.percentile(nz, 1) if nz.size else 1e-3  # linear region around 0

normW = mcolors.SymLogNorm(linthresh=linthresh, vmin=-vmax, vmax=vmax, base=10)

fig, ax = plt.subplots(2, 2, figsize=(14, 12))
im0 = ax[0,0].imshow(C, aspect='auto');                 ax[0,0].set_title("Binary connectivity C")
im1 = ax[0,1].imshow(W, aspect='auto', norm=normW);     ax[0,1].set_title("Weights W (sym-log)")
im2 = ax[1,0].imshow(delays, aspect='auto');            ax[1,0].set_title("Delays (ms)")
im3 = ax[1,1].imshow(D, aspect='auto');                 ax[1,1].set_title("Distance matrix D")

for im, a in zip((im0, im1, im2, im3), ax.ravel()):
    fig.colorbar(im, ax=a, fraction=0.046, pad=0.04)

plt.tight_layout()
plt.show()
```

```python
def build_psp_kernel(dt_ms=1, t_rise_ms=2.0, t_fall_ms=20.0, T_eps_ms=100.0, K=1.0, normalize=None):
    """
    ε(t) = K (e^{-t/τ_fall} - e^{-t/τ_rise}) for 0 <= t <= T_eps, else 0
    normalize: None | 'peak' (max=1) | 'area' (∫ε dt = 1 in ms-units)
    """
    
    t = np.arange(0.0, T_eps_ms + dt_ms, dt_ms)
    k = (np.exp(-t / t_fall_ms) - np.exp(-t / t_rise_ms))
    m = np.max(k)
    if m > 0: 
        k = k / m
    
    return k, (1/m)

dt = 1.0  # ms per bin
kernel, K = build_psp_kernel()
print(build_psp_kernel()[1])
```

```python
def filter_spikes_with_psp(S, kernel):
    """
    S: (N, T) spike counts or 0/1 (per neuron, per time bin)
    kernel: (L,)
    returns F: (N, T) = S convolved with kernel, truncated to length T
    """
    Nn, T = S.shape
    L = kernel.size
    n = T + L - 1
    Sf = np.fft.rfft(S, n=n, axis=1)
    Kf = np.fft.rfft(kernel, n=n)
    F_full = np.fft.irfft(Sf * Kf[None, :], n=n, axis=1)
    return F_full[:, :T]

# Example spike trains (Poisson) for demonstration
# rates in Hz; dt in ms => p = rate * dt/1000
rates = np.zeros(N)
rates[is_inp] = 5.0   # inputs
rates[is_exc] = 1.0
rates[is_inh] = 1.0

T_ms = 500
T = int(T_ms / dt)
S = (rng.random((N, T)) < (rates[:, None] * dt / 1000.0)).astype(float)

F = filter_spikes_with_psp(S, kernel)  # filtered presyn traces
```

```python
def postsyn_drive(W, delays_ms, F, dt_ms, C=None):
    """
    W: (N_post, N_pre) weights (signed OK)
    delays_ms: (N_post, N_pre) propagation delays (ms)
    F: (N_pre, T) presyn filtered traces (after conv with ε)
    C: optional bool connectivity mask; if None, inferred as (W != 0)

    Returns U: (N_post, T) summed postsynaptic drive
    """
    if C is None:
        C = (W != 0)

    d_samples = np.rint(delays_ms / dt_ms).astype(int)
    d_samples[~C] = 0  # ensure we ignore non-edges
    dmax = int(d_samples[C].max(initial=0))

    # Bucket connections by delay: Wk[k, i, j] = W[i, j] if delay(i,j)==k else 0
    Wk = np.stack([(W * C) * (d_samples == k) for k in range(dmax + 1)], axis=0)  # (K, N, N)

    # Pre-shifted presyn traces: Fshift[k, j, t] = F[j, t - k] (zero for t<k)
    Npre, T = F.shape
    Fshift = np.zeros((dmax + 1, Npre, T), dtype=F.dtype)
    Fshift[0] = F
    for k in range(1, dmax + 1):
        Fshift[k, :, k:] = F[:, :T - k]

    # U[i, t] = sum_k sum_j Wk[k, i, j] * Fshift[k, j, t]
    U = np.einsum('kij,kjt->it', Wk, Fshift, optimize=True)
    return U

U = postsyn_drive(W, delays, F, dt, C=C)
```

```python
import numpy as np

class OnlinePSPEvent:
    """
    Online double-exp PSP with per-connection delays.
    Assumes W.shape == delays.shape == (N_pre, N_post).
    """
    def __init__(self, W, delays_ms, dt_ms, t_rise_ms=2.0, t_fall_ms=20.0, K=1.4353055037973592, C=None):
        self.W = W if C is None else W * C
        self.dt = float(dt_ms)
        self.K = float(K)
        self.N_pre, self.N_post = self.W.shape

        # discrete decay factors
        self.alpha_r = K*np.exp(-self.dt / float(t_rise_ms))
        self.alpha_f = K*np.exp(-self.dt / float(t_fall_ms))

        # integer delay buckets
        d_samp = np.rint(delays_ms / self.dt).astype(int)
        if C is not None:
            d_samp = np.where(C, d_samp, 0)
        self.L = int(d_samp.max(initial=0)) + 1  # ring size

        # precompute, for each presyn j, groups of postsyn indices by delay
        self.out_by_delay = []
        for j in range(self.N_pre):
            tgt = np.where(self.W[j] != 0.0)[0]
            if tgt.size == 0:
                self.out_by_delay.append({})
                continue
            dj = d_samp[j, tgt]
            groups = {}
            for k in np.unique(dj):
                sel = (dj == k)
                groups[int(k)] = (tgt[sel], self.W[j, tgt[sel]])
            self.out_by_delay.append(groups)

        # ring buffers for increments arriving this step (per postsyn)
        self._inc_f = np.zeros((self.L, self.N_post), float)
        self._inc_r = np.zeros((self.L, self.N_post), float)
        self._p = 0  # ring pointer

        # postsyn states (sum over all inputs)
        self._gf = np.zeros(self.N_post, float)  # "fall" state
        self._gr = np.zeros(self.N_post, float)  # "rise" state

    def step(self, spikes_pre):
        """
        spikes_pre: (N_pre,) counts or 0/1 in this dt bin.
        Returns U_post: (N_post,) total postsyn drive this step.
        """
        # schedule arrivals for spikes this step
        spk_idx = np.nonzero(spikes_pre)[0]
        for j in spk_idx:
            s = float(spikes_pre[j])
            if s == 0: 
                continue
            for k, (idx, wvals) in self.out_by_delay[j].items():
                slot = (self._p + k) % self.L
                self._inc_f[slot, idx] += s * wvals
                self._inc_r[slot, idx] += s * wvals

        # decay + add arrivals due now
        self._gf = self.alpha_f * self._gf + self._inc_f[self._p]
        self._gr = self.alpha_r * self._gr + self._inc_r[self._p]

        # output and advance ring
        U = self.K * (self._gf - self._gr)
        self._inc_f[self._p].fill(0.0)
        self._inc_r[self._p].fill(0.0)
        self._p = (self._p + 1) % self.L
        return U
```

```python
def bernoulli_spike_step(rng, rates_hz, dt_ms):
    p = rates_hz * (dt_ms / 1000.0)
    p = np.clip(p, 0.0, 1.0)
    return (rng.random(rates_hz.shape) < p).astype(int)
```

```python
import numpy as np

def log_ring_buffer(sim, spikes_seq, i_sel=None):
    """
    sim: OnlinePSPEvent instance (already constructed)
    spikes_seq: (N_pre, T) spike counts for T steps
    i_sel: postsyn index to visualize (default = highest in-degree)
    Returns dict with per-step snapshots.
    """
    N_pre, T = spikes_seq.shape
    L = sim.L
    if i_sel is None:
        i_sel = int(np.argmax((sim.W != 0).sum(axis=0)))

    # logs
    arr_raw   = np.zeros((L, T))   # ring slots as-is (rows=slot 0..L-1)
    arr_align = np.zeros((L, T))   # rows = delay-to-go k=0..L-1 relative to *next* step
    gf, gr, U = np.zeros(T), np.zeros(T), np.zeros(T)
    I_now     = np.zeros(T)        # total arrivals applied at this step for i_sel
    p_hist    = np.zeros(T, dtype=int)

    for t in range(T):
        gf_prev = sim._gf.copy()     # save before step
        p_prev  = sim._p

        U_t = sim.step(spikes_seq[:, t])

        # after step: ring pointer advanced and slot p_prev has been cleared
        p_now = sim._p

        # ring snapshots (use _inc_f; same arrivals used for both gf/gr)
        arr_raw[:, t]   = sim._inc_f[:, i_sel]
        # align so row k = arrivals scheduled to happen k steps from *now*
        for k in range(L):
            arr_align[k, t] = sim._inc_f[(p_now + k) % L, i_sel]

        # states & arrivals used now
        gf[t], gr[t], U[t] = sim._gf[i_sel], sim._gr[i_sel], U_t[i_sel]
        # I_now = gf_new - af * gf_prev  (single neuron)
        I_now[t] = gf[t] - sim.alpha_f * gf_prev[i_sel]
        p_hist[t] = p_now

    return {
        "arr_raw": arr_raw, "arr_align": arr_align,
        "gf": gf, "gr": gr, "U": U,
        "I_now": I_now, "p_hist": p_hist, "i_sel": i_sel, "L": sim.L
    }
```

```python
import matplotlib.pyplot as plt

def plot_ring_viz(log, dt_ms=1.0):
    arr_raw   = log["arr_raw"]
    arr_align = log["arr_align"]
    gf, gr, U = log["gf"], log["gr"], log["U"]
    I_now     = log["I_now"]
    p_hist    = log["p_hist"]
    L         = log["L"]
    T         = arr_raw.shape[1]
    t_ms = np.arange(T) * dt_ms

    fig, axes = plt.subplots(1, 2, figsize=(14, 4), constrained_layout=True)

    # (A) Fixed ring slots; overlay current pointer (slot read next)
    im0 = axes[0].imshow(arr_raw, aspect='auto', origin='lower',
                         extent=[0, T*dt_ms, 0, L])
    axes[0].set_title("Ring contents (raw slots)")
    axes[0].set_xlabel("Time (ms)")
    axes[0].set_ylabel("Ring slot index")
    # overlay pointer trajectory as white line
    axes[0].plot(t_ms, p_hist + 0.5, lw=1.2, color='white', alpha=0.9)
    fig.colorbar(im0, ax=axes[0], fraction=0.046, pad=0.04, label="Scheduled increment")

    # (B) Aligned by delay-to-go (k=0 arrives *next* step)
    im1 = axes[1].imshow(arr_align, aspect='auto', origin='lower',
                         extent=[0, T*dt_ms, 0, L])
    axes[1].set_title("Ring contents (aligned by delay-to-go)")
    axes[1].set_xlabel("Time (ms)")
    axes[1].set_ylabel("k steps until arrival")
    fig.colorbar(im1, ax=axes[1], fraction=0.046, pad=0.04, label="Scheduled increment")
    plt.show()

    # States vs output for the same postsyn neuron
    plt.figure(figsize=(12, 3.6))
    plt.plot(t_ms, gf, label="g_f (slow)")
    plt.plot(t_ms, gr, label="g_r (fast)")
    plt.plot(t_ms, U,  label="U = K(g_f - g_r)", linewidth=2, alpha=0.8)
    plt.step(t_ms, I_now, where='post', label="arrivals now (I_t)", alpha=0.7)
    plt.title("Postsynaptic states and drive (same neuron)")
    plt.xlabel("Time (ms)")
    plt.legend(loc="upper right", ncol=4, frameon=False)
    plt.tight_layout(); plt.show()
```

```python
import numpy as np

def discrete_kernel(dt, t_rise, t_fall, T_eps, K=1.0):
    """Exact discrete double-exp samples matching the online update."""
    af = np.exp(-dt / t_fall)
    ar = np.exp(-dt / t_rise)
    L = int(np.round(T_eps / dt)) + 1
    n = np.arange(L)
    return K * (af**n - ar**n)

def shift_right(x, k):
    """Shift array right by k (pad zeros on left), keep same length."""
    if k <= 0: return x.copy()
    y = np.zeros_like(x)
    if k < len(x):
        y[k:] = x[:-k]
    return y

def reference_output(W, delays_ms, dt, spikes, t_rise, t_fall, T_eps, K=1.0):
    """
    Ground-truth by direct discrete convolution with per-edge delays.
    W: (N_pre, N_post)
    delays_ms: (N_pre, N_post)
    spikes: (N_pre, T) spike counts
    returns U_ref: (N_post, T)
    """
    N_pre, N_post = W.shape
    T = spikes.shape[1]
    ker = discrete_kernel(dt, t_rise, t_fall, T_eps, K)
    # per-pre filtered trace via discrete convolution
    F = np.stack([np.convolve(spikes[j], ker)[:T] for j in range(N_pre)], axis=0)  # (N_pre, T)
    d_samp = np.rint(delays_ms / dt).astype(int)
    U = np.zeros((N_post, T), float)
    for i in range(N_post):
        acc = np.zeros(T, float)
        for j in range(N_pre):
            if W[j, i] == 0: 
                continue
            acc += W[j, i] * shift_right(F[j], d_samp[j, i])
        U[i] = acc
    return U
```

```python
def test_single_no_delay():
    dt, T, t_r, t_f, T_eps, K = 1.0, 120, 2.0, 20.0, 100.0, 1.0
    W = np.array([[1.0]])
    D = np.array([[0.0]])
    sim = OnlinePSPEvent(W, D, dt, t_r, t_f, K)
    spikes = np.zeros((1, T)); spikes[0,0] = 1  # one spike at t=0
    U = np.array([sim.step(spikes[:,t]) for t in range(T)]).T  # (1,T)
    ker = discrete_kernel(dt, t_r, t_f, T_eps, K)
    assert np.allclose(U[0, :len(ker)], ker)
    assert np.all(U[0, len(ker):] == 0)
    print("OK: single synapse, no delay")

def test_single_with_delay():
    dt, T, t_r, t_f, T_eps, K = 1.0, 130, 2.0, 20.0, 100.0, 1.0
    delay = 5.0  # ms
    W = np.array([[1.0]])
    D = np.array([[delay]])
    sim = OnlinePSPEvent(W, D, dt, t_r, t_f, K)
    spikes = np.zeros((1, T)); spikes[0,0] = 1
    U = np.array([sim.step(spikes[:,t]) for t in range(T)]).T  # (1,T)
    ker = discrete_kernel(dt, t_r, t_f, T_eps, K)
    expected = shift_right(ker, int(round(delay/dt)))
    assert np.allclose(U[0, :len(expected)], expected)
    print("OK: single synapse, with delay")

def test_superposition_two_spikes():
    dt, T, t_r, t_f, T_eps, K = 1.0, 140, 2.0, 20.0, 100.0, 1.0
    W = np.array([[1.0]])
    D = np.array([[3.0]])  # 3 ms delay
    sim = OnlinePSPEvent(W, D, dt, t_r, t_f, K)
    spikes = np.zeros((1, T))
    spikes[0, 0] = 1
    spikes[0, 7] = 1
    U = np.array([sim.step(spikes[:,t]) for t in range(T)]).T
    ker = discrete_kernel(dt, t_r, t_f, T_eps, K)
    exp0 = shift_right(ker, int(round(3.0/dt)))
    exp1 = shift_right(ker, int(round(3.0/dt)) + 7)
    expected = exp0 + exp1
    assert np.allclose(U[0, :len(expected)], expected)
    print("OK: superposition of two spikes with delay")

# test_single_no_delay()
test_single_with_delay()
test_superposition_two_spikes()
```

```python
def test_small_network_correctness():
    rng = np.random.default_rng(123)
    dt, T = 1.0, 200
    t_r, t_f, T_eps, K = 2.0, 20.0, 100.0, 1.0

    N_pre, N_post = 3, 2
    W = rng.normal(0, 1, size=(N_pre, N_post))
    # include a negative weight explicitly
    W[0, 1] = -abs(W[0, 1])
    D = rng.integers(0, 8, size=(N_pre, N_post)).astype(float)  # 0..7 ms delays

    # sparse spikes
    spikes = np.zeros((N_pre, T), int)
    spikes[0, [0, 15, 40]] = 1
    spikes[1, [7, 9]] = 1
    spikes[2, [3, 80]] = 2  # counts>1 allowed

    # online simulation
    sim = OnlinePSPEvent(W, D, dt, t_r, t_f, K)
    U = np.zeros((N_post, T))
    for t in range(T):
        U[:, t] = sim.step(spikes[:, t])

    # reference by direct convolution
    U_ref = reference_output(W, D, dt, spikes, t_r, t_f, T_eps, K)

    np.testing.assert_allclose(U, U_ref, rtol=1e-12, atol=1e-12)
    print("OK: small network matches direct reference")

test_small_network_correctness()
```

```python
import time

class DenseDelayBucket:
    """Vector baseline: per-neuron PSP state + delay-bucketed dense matvec."""
    def __init__(self, W, delays_ms, dt_ms, t_rise_ms, t_fall_ms, K=1.0, C=None):
        self.W = W if C is None else W * C
        self.dt = float(dt_ms)
        self.K = float(K)
        self.N_pre, self.N_post = self.W.shape
        self.af = np.exp(-self.dt / float(t_fall_ms))
        self.ar = np.exp(-self.dt / float(t_rise_ms))
        d_samp = np.rint(delays_ms / self.dt).astype(int)
        if C is not None:
            d_samp = np.where(C, d_samp, 0)
        self.L = int(d_samp.max(initial=0)) + 1
        # bucket weights by delay: Wk[k] has same shape as W
        self.Wk = [np.where(d_samp == k, self.W, 0.0) for k in range(self.L)]
        # per-pre states
        self.gf = np.zeros(self.N_pre)
        self.gr = np.zeros(self.N_pre)
        # ring buffer of pre traces
        self.FB = np.zeros((self.L, self.N_pre))
        self.p = 0

    def step(self, spikes_pre):
        self.gf = self.af * self.gf + spikes_pre
        self.gr = self.ar * self.gr + spikes_pre
        f = self.K * (self.gf - self.gr)
        self.FB[self.p] = f
        # accumulate delayed contributions
        U = np.zeros(self.N_post)
        for k in range(self.L):
            f_delay = self.FB[(self.p - k) % self.L]
            # out = W^T @ f
            U += self.Wk[k].T @ f_delay
        self.p = (self.p + 1) % self.L
        return U

def benchmark_compare():
    rng = np.random.default_rng(7)
    # problem size (tune as needed)
    N_pre, N_post = 800, 800
    density = 0.02
    T = 1000
    dt, t_r, t_f, T_eps, K = 1.0, 2.0, 20.0, 100.0, 1.0

    # build sparse-ish W and random delays 0..10 ms on existing edges
    W = np.zeros((N_pre, N_post))
    mask = rng.random((N_pre, N_post)) < density
    W[mask] = rng.normal(0.5, 0.2, size=mask.sum()) * rng.choice([1, -1], size=mask.sum())
    delays = np.zeros_like(W) + 0.0
    delays[mask] = rng.integers(0, 11, size=mask.sum()).astype(float)

    # Poisson spikes
    rate_hz = 5.0
    lam = rate_hz * (dt / 1000.0)
    spikes = rng.poisson(lam, size=(N_pre, T))

    # event-driven
    ev = OnlinePSPEvent(W, delays, dt, t_r, t_f, K)
    t0 = time.perf_counter()
    U_ev = np.zeros((N_post, T))
    for t in range(T):
        U_ev[:, t] = ev.step(spikes[:, t])
    t1 = time.perf_counter()

    # dense vectorized
    de = DenseDelayBucket(W, delays, dt, t_r, t_f, K)
    t2 = time.perf_counter()
    U_de = np.zeros((N_post, T))
    for t in range(T):
        U_de[:, t] = de.step(spikes[:, t])
    t3 = time.perf_counter()

    # correctness & timings
    np.testing.assert_allclose(U_ev, U_de, rtol=1e-10, atol=1e-10)
    print(f"OK: outputs match. Event-driven: {t1 - t0:.3f}s; Dense: {t3 - t2:.3f}s")

benchmark_compare()
```
